```python
import pandas as pd
data=pd.read_csv('creditcard.csv')
print(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
#Visualization using a heat map
plt.figure(figsize=(10,5))
sns.heatmap(data=data.corr(),annot=False,linewidths=0.5)
plt.show()
```

```python
sns.histplot(data=data,x='Time',kde=True)
plt.title('Distribuion of time')
plt.show()
```

```python
#number of transactions and frauds
data['Class'].value_counts()
```

```python
normal=data[data.Class == 0]
fraud =data[data.Class == 1]
```

```python
print(normal.shape)
print(fraud.shape)
```

```python
normal_sample=normal.sample(n=492)
print (normal_sample.shape)
```

```python
new_data=pd.concat([normal_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Applying logistic regression model
model = LogisticRegression()
model.fit(X_train, Y_train)
X_train_prediction = model.predict(X_train)
training_data_adatauracy = accuracy_score(X_train_prediction, Y_train)
print('Adatauracy on Training data : ', training_data_adatauracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
